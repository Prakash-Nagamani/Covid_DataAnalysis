```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
plt.rcParams['figure.figsize'] = [25,10]
```

```python
df = pd.read_csv("PB_cases.csv")
df.head(10)
```

```python
def date_month(date):
    date=date.split("-")
    if date[1]=="01":
        return "January"+date[0]
    if date[1]=="02":
        return "February"+date[0]
    if date[1]=="03":
        return "March"+date[0]
    if date[1]=="04":
        return "April"+date[0]
    if date[1]=="05":
        return "May"+date[0]
    if date[1]=="06":
        return "June"+date[0]
    if date[1]=="07":
        return "July"+date[0]
    if date[1]=="08":
        return "August"+date[0]
    if date[1]=="09":
        return "September"+date[0]
    if date[1]=="10":
        return "October"+date[0]
    if date[1]=="11":
        return "November"+date[0]
    if date[1]=="12":
        return "December"+date[0]
```

```python
month = []
for i in df['date']:
    month.append(date_month(i))
df.insert(31,"month",month,True)
new_df = df.copy()
new_df
```

```python
def variant(date):
    date = date.split('-')
    if date[0] == '2020':
        return 'alpha'
    elif date[0] == '2021':
        if (date[1] == '01' or date[1] == '02' or date[1] == '03'):
            return 'alpha'
        elif date[1] == '12':
            return 'omicron'
        else:
            return 'delta'
    elif date[0] == '2022':
        return 'omicron'
```

```python
variants = []
for i in df['date']:
    variants.append(variant(i))
new_df.insert(32,"variant",variants,True)
new_df
```

```python
new_df['rise in cases']= ""
new_df['rise in cases'] = new_df['active_cases'].diff()
dummy= new_df.copy()
dummy.dropna(subset=['rise in cases'],inplace=True)
dummy.drop(columns=['samples_total','cases_total','isolation','discharged_total','deaths_total','above_45_second_vaccinattion_total','eighteen_44_first_vaccination_total','eighteen_44_second_vaccination_total'],inplace=True)
dummy.drop(columns=['oxygen_support_active_patients','critical_care_active_patients','ventilator_support_active_patients','healthcare_first_vaccination_total','frontline_second_vaccination_total','above_45_first_vaccination_total','total_vaccination_today'],inplace=True)
dummy.drop(columns=['frontline_first_vaccination_total','healthcare_second_vaccination_total'],inplace=True)
# increase= dummy[dummy['rise in cases']>0]
# decrease = dummy[dummy['rise in cases']<0]

# increase
dummy
```

```python
def bar_graph(categories,cat_val,xlabel,ylabel,title=""):
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.bar(categories,cat_val,width=0.8)
    plt.title(title)
    plt.show()
```
