```python
def date_month(date):
    date=date.split("-")
    if date[1]=="01":
        return "January"+date[0]
    if date[1]=="02":
        return "February"+date[0]
    if date[1]=="03":
        return "March"+date[0]
    if date[1]=="04":
        return "April"+date[0]
    if date[1]=="05":
        return "May"+date[0]
    if date[1]=="06":
        return "June"+date[0]
    if date[1]=="07":
        return "July"+date[0]
    if date[1]=="08":
        return "August"+date[0]
    if date[1]=="09":
        return "September"+date[0]
    if date[1]=="10":
        return "October"+date[0]
    if date[1]=="11":
        return "November"+date[0]
    if date[1]=="12":
        return "December"+date[0]
```

```python
month = []
for i in df['date']:
    month.append(date_month(i))
month
df.insert(11,"month",month,True)
new_df = df.copy()
```

```python
import datetime 
def findDay(date):
    calendar=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    date=date.replace("-"," ")
    born = datetime.datetime.strptime(date,'%Y %m %d').weekday()
#     print(born)
    return (calendar[born])
```

```python
day = []
for i in new_df['date']:
    day.append(findDay(i))
new_df.insert(12,"day",day,True)
```

```python
# latest_cases.plot(x=latest_cases['district'].values,y="active_cases",kind='pie')
# plt.xlabel()
def piechart(categories,cat_val,title=""):
    labels=categories
    values=cat_val
#     explode = (0,0.1)
    plt.pie(values,labels=values,counterclock=False,autopct="%.2f", shadow=True)
    plt.title(title)
    plt.legend(labels,loc=3)
    plt.rcParams['figure.figsize'] = [10,50]
    plt.show()
```

```python
def bar_graph(categories,cat_val,xlabel,ylabel,title=""):
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.bar(categories,cat_val,width=0.8)
    plt.title(title)
    plt.show()
```
