```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import random
%matplotlib inline
plt.rcParams['figure.figsize'] = [30,12]
```

```python
df=pd.read_csv("D:\programs\datathon\datasets\case_info\TN_district_details.csv")
df
```

```python
dfgrouped=df.groupby(by=["district"])
```

```python
districts=[]
for i in list(dfgrouped.groups.keys()):
    if i == "grand total":
        continue
    districts.append(dfgrouped.get_group(i))
```

```python
lsi=[]
for i in range(len(districts)):
    lsi.append(districts[i]["total_cases_till_today"].max())
    print(list(districts[i]["district"])[0])
print(lsi)
```

```python
plt.rcParams['figure.figsize'] = [30,9]
plt.title("districts")
plt.ticklabel_format(style='plain')
for i in range(len(districts)):
    plt.plot(districts[i]["date"],districts[i]["new_cases_today"])
plt.legend(list(dfgrouped.groups.keys()))
```

```python
def random_color():
  rgbl=[255,0,0]
  random.shuffle(rgbl)
  return tuple(rgbl)
```

```python
def piechart(categories,cat_val,n,title=""):
    labels=categories
    values=cat_val
    explode = [0.1]*n
    plt.pie(values,colors=["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
             for i in range(n+1)],labels=values,explode=explode,counterclock=False,autopct="%.2f", shadow=True)
    plt.title(title)
    plt.legend(labels,loc=3)
    plt.rcParams['figure.figsize'] = [10,50]
    plt.rcParams["figure.autolayout"] = True
    plt.show()
```

Description: Pie chart depicting districts with mean no of +ve cases greater than the mean of +ve cases of all districts

```python
discmean=[]
meannames=[]
actualkeys=[]
for i in range(len(districts)):
    discmean.append(list(districts[i]["total_cases_till_today"])[-1])
piechart(list(dfgrouped.groups.keys()),discmean,len(discmean),"district_wise total +ve cases")
```

```python
import datetime 
def findDay(date):
    calendar=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    date=date.replace("-"," ")
    born = datetime.datetime.strptime(date, '%d %m %Y').weekday()
    return (calendar[born])
```

```python

for j in range(len(districts)):
    day=[]
    for i in districts[j]["date"]:
        day.append(findDay(i))
    districts[j].insert(1,"day",day,True)
```

```python
districts[3]
```

```python
plt.rcParams['figure.figsize'] = [30,12]
dfgroupday=[]
for i in range(len(districts)):
    plt.figure(i)
    plt.title(list(districts[i]["district"])[0])
    dfgroupday.append(districts[i].groupby(by="day").agg("mean"))
    plt.bar(["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"],dfgroupday[i]["new_cases_today"])
```

## TN incoming travellers analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [30,12]
```

```python
df=pd.read_csv("D:\\programs\\datathon\\datasets\\travel\\TN_incoming_people_till_yesterday.csv")
df
```

```python
uniqdates=[]
dfgrouped=df.groupby(by=["date"])
for i in list(df["date"].unique()):
    x=dfgrouped.get_group(i)
    x=x[x.travel_mode != "grand total"]
    uniqdates.append(x)
uniqdates
```

```python
sumravelmod=[]
lissum=[]
for i in range(len(uniqdates)):
    sumravelmod.append(uniqdates[i].groupby("date").sum().reset_index())
    lissum.append(sumravelmod[i]["positive_cases"][0])
sumravelmod[0]["positive_cases"][0]
```

```python
def bar_graph(categories,cat_val,xlabel,ylabel,title=""):
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.bar(categories,cat_val,width=0.8)
    plt.title(title)
    plt.show()
```

Description: graph plotting positives cases against date for each mode of transport.

```python
plt.title("modes positive cases")
plt.xlabel("date")
plt.ylabel("positive cases")
plt.plot(df["date"].unique(),lissum,color="blue")
```

```python
# This is the code of linear_regression model also.........           
def linear_correlation(x,y):
    meanx=sum(x.values)/len(x)
    meany=sum(y.values)/len(y)
    if len(x.values)==len(y.values):
        num,den=0,0
        #numerator calculation
        a=0
        for i in range(len(x)):
            a+=x.values[i]*y.values[i]
            #         print(a)
        num=a*len(x)-(sum(x.values)*sum(y.values))
    
        #denominator calculation
        ax,ay=0,0
        for i in range(len(x)):
                ax+=x.values[i]**2
                ay+=y.values[i]**2 
        p1=(len(x)*ax-(sum(x.values)**2))**0.5
        p2=(len(y)*ay-(sum(y.values)**2))**0.5
        den=p1*p2
        #print(num,den)
        return num/den
    else:
        raise Exception("Invalid Dimension")
```
