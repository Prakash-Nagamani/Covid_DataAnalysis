```python
def variant(date):
    date = date.split('-')
    if date[2] == '2020':
        return 'alpha'
    elif date[2] == '2021':
        if (date[1] == '01' or date[1] == '02' or date[1] == '03'):
            return 'alpha'
        elif date[1] == '12':
            return 'omicron'
        else:
            return 'delta'
    elif date[2] == '2022':
        return 'omicron'
```

```python
variantlis=[]
for i in list(df["date"]):
    variantlis.append(variant(i))
df.insert(2,"variant",variantlis)
df
```

description: Graph showing affect of each variant

```python
variants=["alpha","delta","omicron"]
for i in variants:
    x=df.groupby(by="variant").get_group(i)
    plt.plot(x["date"],x["cases_new"])
```

Description: Below graph depicts the amount of deaths per day due to each variant.

```python

variants=["alpha","delta","omicron"]
for i in variants:
    x=df.groupby(by="variant").get_group(i)
    plt.plot(x["date"],x["deaths_new"])
```

```python
dfdistrict=pd.read_csv("D:\programs\datathon\datasets\case_info\TG_district_cases_info.csv")
dfdistrict
```

```python
districts=list(dfdistrict["district"].unique())
dfmean=dfdistrict.groupby(by="district").agg("mean")
dfmean
```

```python
def piechart(categories,cat_val,n,title=""):
    labels=categories
    values=cat_val
    explode = [0.1]*n
    plt.pie(values,colors=["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
             for i in range(n+1)],labels=values,explode=explode,counterclock=False,autopct="%.2f", shadow=True)
    plt.title(title)
    plt.legend(labels,loc=3)
    plt.rcParams['figure.figsize'] = [10,50]
    plt.rcParams["figure.autolayout"] = True
    plt.show()
```

```python
piechart(districts,dfmean["cases_new"],len(districts),"District wise cases")
```
